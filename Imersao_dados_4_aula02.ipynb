#Desafio 3

```python
media_Metragem=dados_vendas_n.groupby(["Bairro"]).mean("Valor_anuncio_float")
media_Metragem["Valor_metro_quadrado"] = pd.to_numeric(media_Metragem["Valor_anuncio_float"]) / pd.to_numeric(media_Metragem["Metragem"])
```

```python
media_Metragem
```

```python
plt.figure(figsize=(20,5))
media_Metragem["Valor_metro_quadrado"].plot.line()
```

#Desafio 4

```python
sns.set()
plt.figure(figsize=(20, 8))
ax = sns.histplot(data = media_Metragem, x = "Valor_metro_quadrado",color='r')
ax.set_title("Valor metro Quadrado X Bairros")
ax.set_ylabel("Qtde Bairros")
ax.set_xlabel("Valor metro quadrado")
plt.xlim((-0.1, 20000))
plt.show()
```
